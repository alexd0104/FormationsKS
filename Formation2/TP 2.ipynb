```python
import pandas as pd 
import xgboost as xgb
```

```python
# On importe la dataset

df = pd.read_csv("/kaggle/input/bostonhoustingmlnd/housing.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
y = df['MEDV'] # On mets la cible dans une colonne à part
X = df.drop (columns = 'MEDV') # On enlève la cible des données d'entrainement
```

```python
model = xgb.XGBRegressor()
```

```python
model.fit(X, y) # Le modèle trouve la meilleure relation pour passer de X à y 
```

```python
model.predict(X) # On calcul les valeurs les plus probables de MEDV selon les autres colonnes 
```
